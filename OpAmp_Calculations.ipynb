# For Potential Op Amp Setup with use of the Load Cell
Running through several different Resistors to determine what level of gain needed to get good voltage readings



```python
import numpy as np
from matplotlib import pyplot as plt

def Vout(Vin,G):
    return Vin*G

def G(Rg):
    return 5 + 80000/Rg

dVin = .001                     # From datasheet, output is 1 mV/V
Vin_max = 5*dVin                # Power from arduino 5V
Vin = np.arange(0,Vin_max,dVin)        #V (arbitary number just to see size of gain)
Rg = [75, 100, 150, 220, 330, 470,680,1000,2200,3000]

plt.figure(figsize=(6,4),dpi=150)
for Rg in Rg:
    plt.plot(Vin,Vout(Vin,G(Rg)),label='{} $\Omega$'.format(Rg))
    print('Resistor: {}, Vout: {}, Gain: {}'.format(Rg,Vout(Vin,G(Rg)),G(Rg)))
    plt.xlabel('Vin (V)')
    plt.ylabel('Vout (V)')
    plt.title('Vout v Vin of Op amp for different Rg')
    plt.legend()


```
